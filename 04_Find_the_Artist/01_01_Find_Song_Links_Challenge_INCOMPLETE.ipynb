Make a URL request

```python
url = "https://www.lyrics.com/artist/Helge-Schneider/748954"
param_dict = {'db':'pubmed', 'term':'escherichia', 'rettype':'uilist'}

r_helge = requests.get(url, params=param_dict)
```

Decipher URL formatting with regular expressions

```python
# <a href="/lyric/10047067/Helge+Schneider/Texas">Texas</a>
```

```python
pattern = "lyric/\d+/Helge\+Schneider/[^\"]+"
```

Find all songs matching RE pattern

```python
songs_helge = re.findall(pattern, r_helge.text, re.IGNORECASE)
```

```python
songs_helge
```

Loop that goes through all song URLs that were collected previously and construct a complete URL

```python
for song in range(len(songs_helge)):
    songs_helge[song] = 'https://www.lyrics.com/' + songs_helge[song]
```

```python
songs_helge
```

Generate a unique file name (using the song name or a number)

```python
pattern = "lyric/\d+/Helge\+Schneider/([^\"]+)"

song_names_helge = re.findall(pattern, r_helge.text)
```

```python
song_names_helge
```

```python
song_names_helge[0]
```

Replace special characters to increase readability

```python
for song in range(len(song_names_helge)):
    song_names_helge[song] = song_names_helge[song].replace("%27", "_")
    song_names_helge[song] = song_names_helge[song].replace("+", "_")
```

If you'd want to do it manually, you could use nested for loops with lists, e. g. like this:

```python
for song in range(len(song_names_helge)):
    song_as_list = list(song_names_helge[song])
    for letter in range(len(song_as_list)):
        if (song_as_list[letter] == '%') or (song_as_list[letter] == '+'):
            song_as_list[letter] = '_'
        elif (song_as_list[letter] == '2') or (song_as_list[letter] == '7'):
            song_as_list[letter] = ''
    song_names_helge[song] = ''.join(song_as_list)
```

Check, if song names were converted correctly

```python
print(song_names_helge) 
```

Loop through the newly created links and save each of them as an individual file with its corresponding file name

```python
for song in range(len(songs_helge)):
    url = songs_helge[song]
    r_helge = requests.get(url)
    file_name = '_RES/helge/Helge_Schneider_' + str(song_names_helge[song] + '.txt')
    with open (file_name, 'w') as file:
        file.write(r_helge.text)
```

#### Beautiful Soup

```python
url = 'https://www.lyrics.com/artist/Foo-Fighters/144725'

response = requests.get(url)
```

```python
fighters_soup = BeautifulSoup(fighters_html)
```

```python
fighters_soup
```

```python
fighters_soup.body.find_all(attrs={'class':'tal qx'})
```

```python
fighters_soup.body.find_all(attrs={'class':'tal qx'})[0].text
```

```python
all_songs_foo_fighters = []
all_songs_and_links_foo_fighters = []

for index in range(len(fighters_soup.body.find_all(attrs={'class':'tal qx'}))):
    song = fighters_soup.body.find_all(attrs={'class':'tal qx'})[index].text
    song_and_link = str(fighters_soup.body.find_all(attrs={'class':'tal qx'})[index].text) +\
                            ': https://www.lyrics.com' + \
                            fighters_soup.body.find_all(attrs={'class':'tal qx'})[index].find('a').get('href')
    if (song in all_songs_foo_fighters):
        continue
    elif ']' in song:
        continue
    else:
        all_songs_foo_fighters.append(song)        
        all_songs_and_links_foo_fighters.append(song_and_link)
        
all_songs_and_links_foo_fighters.sort()
```
