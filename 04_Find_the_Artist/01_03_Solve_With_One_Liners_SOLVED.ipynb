```python
import requests
import re
from collections import Counter
```

```python
# 1. retrieve the Berlin Wikipedia page
response = requests.get("https://en.wikipedia.org/wiki/Berlin")
```

```python
# 2. check the status code
print(response.status_code)
```

```python
# 3. store the html contents of the page
html = response.text
```

```python
# 4. write the html contents to a file
with open ("_RES/berlin.html", "w") as f: f.write(html)
```

```python
print(html)
```

```python
# 5. remove all html tags
text = re.sub('\<.+\>', '', html)
```

```python
text
```

```python
# 6. extract words
words = re.findall('\w+', text)
```

```python
words
```

```python
# 7. remove words with less than 6 characters
s = " ";
words = s.join(words)
```

```python
words
```

```python
long_words = re.findall('\w{6,}', words)
```

```python
long_words
```

```python
# 8. create a Counter object from the long words
c = Counter(long_words)
```

```python
# 9. print how often the word 'Berlin' occurs
print(c['Berlin'])
```

```python
# 10. print the 10 most common words
print(c.most_common(10))
```
