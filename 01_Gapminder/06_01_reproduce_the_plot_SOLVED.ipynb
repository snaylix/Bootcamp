```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-10, 10, num=41)
x
type(x)
```

#### Vectorization

#### With Numpy we have to write fewer for loops

```python
y = x ** 2
# alt np.power(x, 2)
```

```python
len(x), len(y)
```

```python
x.shape, y.shape
```

```python
plt.figure(figsize=(14, 7)) # start new figure, figsize in inches

plt.subplot(1, 2, 1)
plt.plot(x, y, color="forestgreen") # 140 color names which can be found on the Internet
plt.subplot(1, 2, 2)
plt.plot(x, y, '.', color=(1.0, 0, 0.75), label='parabola') # RGB tuple
plt.plot(x, np.sin(x) * 50 + 10, color="forestgreen")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial $y = x^2$')
plt.legend() # <-- alt: put labels here

plt.savefig('_RES/parabola.png', dpi=150) # dots per inch: 1500x 1050 pixels
```
